```python
df_fs = df.copy()
df_fs
```

```python
print(
df_fs.select_dtypes(
include='object'
).columns
)
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df_fs.copy()

le_dict = {}

for col in df_encoded.select_dtypes(include='object').columns:

    le = LabelEncoder()

    df_encoded[col] = le.fit_transform(
        df_encoded[col]
    )

    le_dict[col] = le

print("Encoding Completed")
```

```python
df_encoded.head()
```

```python
# ==========================
# FINAL INPUT FEATURES
# ==========================

X = df_encoded[

[
'age',
'gender',
'course',
'year',

'daily_study_hours',
'daily_sleep_hours',
'screen_time_hours',
'physical_activity_hours',

'attendance_percentage',
'cgpa',

'academic_pressure_score',
'financial_stress_score',

'social_support_score',
'family_support_score',
'peer_relationship_score',

'career_uncertainty_score',

'sleep_quality',
'internet_quality',

'caffeine_intake_per_day',
'part_time_job_hours',
'extracurricular_hours'
]

]

# ==========================
# FINAL TARGETS
# ==========================

Y = df_encoded[
[
'stress_score',
'anxiety_score',
'depression_score',
'burnout_score',
'sleep_disorder_score',
'wellness_score'
]
]

# ==========================
# CHECK SHAPE
# ==========================

print("Input Shape :", X.shape)
print("Output Shape :", Y.shape)

print("\nInput Features:\n")
print(X.columns.tolist())

print("\nTarget Columns:\n")
print(Y.columns.tolist())
```

```python
X.describe()
```

```python
for col in Y.columns:
    print("\n", col)
    print(Y[col].value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.20,
    random_state=42
)

print("X Train :", X_train.shape)
print("X Test  :", X_test.shape)

print("Y Train :", Y_train.shape)
print("Y Test  :", Y_test.shape)
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

model = MultiOutputRegressor(rf)

model.fit(
    X_train,
    Y_train
)

print("Regression Model Trained ✅")
```

```python
Y_pred = model.predict(X_test)

print("Prediction Shape :", Y_pred.shape)

print(Y_pred[:5])
```

```python
from sklearn.metrics import accuracy_score

target_names = Y.columns

for i, target in enumerate(target_names):

    acc = accuracy_score(
        Y_test.iloc[:, i],
        Y_pred[:, i]
    )

    print(f"{target} Accuracy : {acc:.4f}")
```

```python
from sklearn.metrics import r2_score

for i,col in enumerate(Y.columns):

    r2 = r2_score(
        Y_test.iloc[:,i],
        Y_pred[:,i]
    )

    print(f"{col} R2 Score : {r2:.3f}")
```

```python
from xgboost import XGBRegressor
from sklearn.multioutput import MultiOutputRegressor

xgb = XGBRegressor(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    random_state=42
)

model_xgb = MultiOutputRegressor(xgb)

model_xgb.fit(
    X_train,
    Y_train
)

print("XGBoost Training Complete ✅")
```

```python
Y_pred_xgb = model_xgb.predict(X_test)
Y_pred_xgb
```

```python
from sklearn.metrics import r2_score

for i,col in enumerate(Y.columns):

    r2 = r2_score(
        Y_test.iloc[:,i],
        Y_pred_xgb[:,i]
    )

    print(f"{col} R2 Score : {r2:.3f}")
```
